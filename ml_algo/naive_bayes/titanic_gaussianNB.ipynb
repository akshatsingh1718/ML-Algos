```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
df = titanic.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)
df.head()
```

```python
gender_dummies = pd.get_dummies(df.Sex)
gender_dummies.head()
```

```python
df = pd.concat([df, gender_dummies], axis='columns')
df.head()
```

#### I am dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female

```python
df.drop(['Sex', 'male'], axis=1, inplace=True)
df.head()
```

#### Checking if any column contains Nan value

```python
df.columns[df.isna().any()]
```

#### Filling Nan values in Age cloumn

```python
df.Age = df.Age.fillna(df.Age.mean())
df.head()
```

```python
input = df.drop('Survived', axis=1)
target = df.Survived
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input, target, test_size = 0.3)
```

#### importing GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
model.predict_proba(x_test)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(GaussianNB(), x_train, y_train, cv=5)
score
```
